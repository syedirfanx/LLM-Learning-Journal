### Load Word2Vec (Google News 300)
*Loading Pre-trained Word2Vec Model (Google News, 300 Dimensions)*

```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")
```

### Accessing Word Embedding
*The numerical vector representation of a specific word from the pre-trained model to understand its semantic meaninng*

```python
word_vectors=model

print(word_vectors['Microsoft'])
```

### Word Analogy Example
*Performing vector arithmetic to explore relationships between words using most_similar*

```python
print(word_vectors.most_similar(positive=['king', 'woman'], negative=['man'], topn=3))
```

```python
print(word_vectors.most_similar(positive=['Paris', 'Italy'], negative=['France'], topn=3))
```

```python
print(word_vectors.most_similar(positive=['Tokyo', 'Cairo'], negative=['Japan'], topn=3))
```
